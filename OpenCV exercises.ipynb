```python
# Load two images
img1 = cv.imread('messi5.jpg')
img2 = cv.imread('opencv_logo.png')
# I want to put logo on top-left corner, So I create a ROI
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols ]

# Now create a mask of logo and create its inverse mask also
img2gray = cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
cv.imshow('im2gray',img2gray)
ret, mask = cv.threshold(img2gray, 10, 255, cv.THRESH_BINARY)
cv.imshow('m',mask)
mask_inv = cv.bitwise_not(mask)
cv.imshow('mi',mask_inv)

# Now black-out the area of logo in ROI
img1_bg = cv.bitwise_and(roi,roi,mask = mask_inv)
cv.imshow('1st',img1_bg)
# Take only region of logo from logo image.
img2_fg = cv.bitwise_and(img2,img2,mask = mask)
cv.imshow('2nd',img2_fg)
# Put logo in ROI and modify the main image
dst = cv.add(img1_bg,img2_fg)
cv.imshow('Add 1 + 2',dst)
img1[0:rows, 0:cols ] = dst


cv.imshow('res',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2;import numpy as np;import matplotlib.pyplot as plt

traind=np.random.randint(0,100,(50,2)).astype(np.float32)
labels=np.random.randint(0,2,(50,1)).astype(np.float32)

red=traind[labels.ravel()==0]
plt.scatter(red[:,0],red[:,1],80,'r','^')

blue=traind[labels.ravel()==1]
plt.scatter(blue[:,0],blue[:,1],80,'b','s')



#adding a newcomer
newcomer=np.random.randint(0,100,(1,2)).astype(np.float32)
plt.scatter(newcomer[:,0],newcomer[:,1],80,'g','o')
plt.show()
#finding nearest...
knn=cv2.ml.KNearest_create()
knn = cv2.ml.KNearest_create()
knn.train(traind,cv2.ml.ROW_SAMPLE,labels)
ret, results, neighbours, dist = knn.findNearest(newcomer, 3)

print(dist)
print(neighbours)
print(results)
```
